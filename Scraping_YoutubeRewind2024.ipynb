```python
from googleapiclient.discovery import build
from langdetect import detect
import pandas as pd

# API key
api_key = 'AIzaSyCVLpah-8p0O7wTGhliilJyZwPmFXUBvMQ'
youtube = build('youtube', 'v3', developerKey=api_key)

def get_video_comments(video_id, max_results=100):
    comments = []
    next_page_token = None

    while len(comments) < max_results:
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            textFormat="plainText",
            maxResults=min(100, max_results - len(comments)),
            pageToken=next_page_token
        ).execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            published_at = item['snippet']['topLevelComment']['snippet']['publishedAt']
            try:
                language = detect(comment)
            except:
                language = "unknown"
            comments.append({"comment": comment, "publishedAt": published_at, "language": language})

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return comments[:max_results]

# ID video
video_id = 'ff_EnPWRrII'

comments = get_video_comments(video_id, max_results=15000)

for idx, comment_data in enumerate(comments, 1):
    print(f"{idx}. {comment_data['comment']} (Published at: {comment_data['publishedAt']}, Language: {comment_data['language']})")
```
